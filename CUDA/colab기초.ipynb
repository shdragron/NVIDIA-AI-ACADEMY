```python
%%cuda
#include <iostream>
__global__
void add(int a, int b, int c) {
    c = a + b;
    printf("Hello World %d \n", c);
}

int main() {
    std::cout << "Hello World\n";
    add<<<1,1>>>(1,2,3);
    cudaDeviceSynchronize();
    return 0;
}
```
